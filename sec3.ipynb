```python
from Bio.Blast import NCBIWWW

# Define protein sequence
sequence = "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHG"

# Run BLAST search with text output format
result_handle = NCBIWWW.qblast("blastp", "nr", sequence, format_type="Text")

# Save result to file
with open("blast_results.txt", "w") as file:
    file.write(result_handle.read())
```

qblast("blastp", "nr", sequence):
"blastp" → يستخدم BLASTP (للبحث عن البروتينات المتشابهة)

"nr" → البحث يتم في قاعدة بيانات "nr" (Non-Redundant Protein Database) التي تحتوي على جميع تسلسلات البروتين الفريدة من مختلف قواعد البيانات
sequence → التسلسل الذي نريد البحث عنه.
✅ لماذا نستخدم "nr"؟

"nr" (Non-Redundant) قاعدة بيانات ضخمة تحتوي على كل البروتينات الفريدة من مصادر متعددة، مما يجعل البحث أكثر شمولًا وموثوقية.
إذا استخدمنا قاعدة بيانات أخرى (مثل swissprot)، فستكون النتائج محدودة بمصادر معينة

```python
from Bio.Blast import NCBIWWW, NCBIXML

# Run BLAST search
result_handle = NCBIWWW.qblast("blastp", "nr", sequence)
blast_record = NCBIXML.read(result_handle)

# Print top 3 matches
for alignment in blast_record.alignments[:3]:
    print(f"Match: {alignment.title}")
    for hsp in alignment.hsps:
        print(f"Score: {hsp.score}, E-value: {hsp.expect}")
```

 طباعة أفضل 3 تطابقات من نتائج البحث

blast_record.alignments[:3] → نحصل على أول 3 تطابقات

alignment.title → اسم البروتين الذي تم العثور عليه

hsp.score → درجة التطابق (كلما زادت، كان التطابق أفضل)

hsp.expect → قيمة E-value التي تشير إلى احتمالية العثور على هذا التطابق بالصدفة (كلما كانت أصغر، كان التطابق أكثر دقة)
