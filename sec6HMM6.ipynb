 2: The HMM Algorithm for Protein Structure Prediction
Hidden Markov Model (HMM):

Predicts protein secondary structures based on amino acid sequences.

Goal: Identify whether parts of the sequence form:

Helix (H) ðŸŒ€

Sheet (S) ðŸ“„

Key Inputs:

States: Helix (H), Sheet (S)

Observations: Amino acids (A, C, G, T)

Transition Probabilities: Probability of moving from one state to another (e.g., H to H, S to H)

Emission Probabilities: Probability of an amino acid occurring in a given state

```python
import numpy as np
from Bio import motifs

# Define a simple set of states (H: Helix, S: Sheet)
states = ['H', 'S']

# Define possible amino acid residues as observations
observations = ['A', 'C', 'G', 'T']

# Define transition probabilities between states
# From state H to H, H to S, S to H, and S to S
transition_probs = {
    'H': {'H': 0.7, 'S': 0.3},
    'S': {'H': 0.4, 'S': 0.6}
}

# Define emission probabilities for each state (H and S)
emission_probs = {
    'H': {'A': 0.2, 'C': 0.3, 'G': 0.4, 'T': 0.1},
    'S': {'A': 0.4, 'C': 0.2, 'G': 0.2, 'T': 0.2}
}

# Define initial state probabilities
start_probs = {'H': 0.5, 'S': 0.5}

# Example sequence of observations (protein sequence)
sequence = ['A', 'C', 'G', 'T', 'A']

```

 3: Viterbi Algorithm Output - Results & Interpretation
Viterbi Algorithm Execution:

Most Likely Hidden Path: Identifies which state (H or S) each amino acid in the sequence belongs to.

Example Output:
['S', 'H', 'H', 'S', 'S']

A = Beta Sheet (S)

C = Alpha Helix (H)

G = Alpha Helix (H)

T = Beta Sheet (S)

A = Beta Sheet (S)

Likelihood of the Sequence:

Probability of the best hidden state sequence:
0.0000968

This low number indicates the probability of the observed sequence happening under the given model, calculated using logarithms.

Applications:

Protein Structure Prediction

Understanding Protein Function & Folding

```python
def viterbi(sequence, states, start_probs, transition_probs, emission_probs):
    # Initialize the dynamic programming table
    dp = {}
    backpointer = {}

    # Initialize base cases (first observation)
    for state in states:
        dp[state] = {}
        dp[state][0] = np.log(start_probs[state]) + np.log(emission_probs[state][sequence[0]])
        backpointer[state] = {}

    # Fill in the DP table
    for t in range(1, len(sequence)):
        for state in states:
            max_prob = -np.inf
            prev_state = None
            for prev in states:
                prob = dp[prev][t-1] + np.log(transition_probs[prev][state]) + np.log(emission_probs[state][sequence[t]])
                if prob > max_prob:
                    max_prob = prob
                    prev_state = prev
            dp[state][t] = max_prob
            backpointer[state][t] = prev_state

    # Backtrack to find the best path
    best_path = []
    max_final_state = max(states, key=lambda state: dp[state][len(sequence)-1])
    best_path.append(max_final_state)

    for t in range(len(sequence)-1, 0, -1):
        best_path.insert(0, backpointer[best_path[0]][t])

    return best_path, dp[max_final_state][len(sequence)-1]

# Run Viterbi to find the most likely sequence of states
best_path, likelihood = viterbi(sequence, states, start_probs, transition_probs, emission_probs)

print(f"Most likely hidden path: {best_path}")
print(f"Likelihood of the sequence: {np.exp(likelihood)}")
```
